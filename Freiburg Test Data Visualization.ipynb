```python
import matplotlib.pyplot as plt
import cv2
```

```python
def display_images_side_by_side(img1, img2, img3):
    # Create a figure with 1 row and 3 columns
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Display each image in a different subplot
    axes[0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
    axes[0].axis('off')  # Turn off axis
    axes[0].set_title("GT Depthmap")

    axes[1].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    axes[1].axis('off')  # Turn off axis
    axes[1].set_title("Predicted Depthmap")

    axes[2].imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
    axes[2].axis('off')  # Turn off axis
    axes[2].set_title("Depthmap overlayed with RGB Image")

    # Display the images
    plt.tight_layout()
    plt.show()
```

```python
img1 = cv2.imread("/home/ahmad/courses/cuda_lab/Project/ThermalVision3D/freiburg_result_images/pair_0_11_gt.png")
img2 = cv2.imread("/home/ahmad/courses/cuda_lab/Project/ThermalVision3D/freiburg_result_images/pair_0_11_depth.png")
img3 = cv2.imread("/home/ahmad/courses/cuda_lab/Project/ThermalVision3D/freiburg_result_images/pair_0_11_overlay.png")
display_images_side_by_side(img1, img2, img3)
```
